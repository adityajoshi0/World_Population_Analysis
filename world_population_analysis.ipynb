```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import seaborn as sns
sns.histplot(df["2022 Population"], bins=10, kde=True)
plt.title("distribution of 2022 population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("Continent")["2022 Population"].mean().sort_values(ascending=False).tail(2)
```

```python
top_10_countries = df.sort_values(by= "2022 Population", ascending = False).head(10)
```

```python
top_10_countries
```

```python
sns.barplot(x="2022 Population", y="Country/Territory", data=top_10_countries)
plt.title("top 10 countries by Population in 2022")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
df.columns
```

```python
df.rename(columns = {"Density (per km²)": "density (km2)"},inplace = True)
```

```python
sns.scatterplot(x="2022 Population", y="density (km2)", data=df)
plt.title("Population and Density Scatter Plot")
plt.xlabel("2022 Population")
plt.ylabel("Density")
plt.show()
```

```python
df.columns
```

```python
df.rename(columns = {"Area (km²)": "area (km2)"},inplace = True)
```

```python
df.groupby("Continent")["area (km2)"].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="Continent", y="density (km2)", data=df, palette= "viridis")
plt.title("Population Density Distribution Across Continents")
plt.xlabel("Continents")
plt.ylabel("Population Density")
plt.show()
```

```python
sns.boxplot(x="Continent", y="density (km2)", data=df, palette= "muted", showfliers=False)
plt.title("Population Density Distribution Across Continents(without outliers)")
plt.xlabel("Continents")
plt.ylabel("Population Density")
plt.show()
```
